# All Arguments (*args)

```python
def productdetails(id, *args):
    print(id)
    for data in args:
        print(data)
productdetails(1, 'Name', 'price')
```

# All Keyword Arguments

```python
def productdetails(**kwargs):
    # print args
    for key, val in kwargs.items():
        print(f'{key}: {val}')
productdetails(id=1, Name='Sam', price='Rs 200')
```

```python
def add(a, b):
    return f'addition of {a} and {b}:{a+b}'
def sub(a, b):
    return f'subtract of {a} and {b}: {a-b}'
def mul(a, b):
    return f'multiply of {a} and {b}: {a*b}'
def division(a, b):
    return f'division of {a} and {b}: {a/b}'

    
```

```python
def calculator(val1, val2):
    return add(val1, val2), sub(val1, val2), mul(val1, val2), division(val1, val2)
```

```python
calculator(3, 5)
```

```python
def calculator(val1, val2):
    val_add=add(val1, val2)
    val_sub=sub(val1, val2)
    val_mul=mul(val1, val2)
    val_division=division(val1, val2)

    print(val_add)
    print(val_sub)
    print(val_mul)
    print(val_division)
```

```python
calculator(3, 5)
```

```python
def calculator(val1, val2):
    choice = input('enter your choice(+, -, *, /):')
    if choice == '+':
        val_add=add(val1, val2)
        print(val_add)
    elif choice =='*':
        val_sub=sub(val1, val2)
        print(val_sub)
    elif choice=='*':
        val_mul=mul(val1, val2)
        print(val_mul)
    elif choice=='/':
        val_division=division(val1, val2)
        print(val_division)

   
```

```python
calculator(3, 5)
```
