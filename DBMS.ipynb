```python
import sqlite3
```

```python
#connects python with database file
conn = sqlite3.connect('ecommerce.db')
```

```python
# executes code from python to database
cursor = conn.cursor()
```

```python
try:
    cursor.execute('''
        CREATE TABLE customers(
            customer_id integer PRIMARY KEY AUTOINCREMENT, --Primary key jaslai banaincha tei keyword use hune ho
            customer_name text,
            customer_phone text,
            customer_email text,
            customer_address text
         );
     ''')
    print('Customer table created')
except sqlite3.OperationalError as e:
    print(e)
```

# Create Staff Table

```python
try:
    cursor.execute('''
        Create table staff(
              staff_id integer primary key autoincrement,
              staff_name text,
              staff_email,
              staff_number text,
              staff_salary real,
              position text
        );
    ''')
    print('Staff table created')
except sqlite3.OperationalError as e:
    print(e)
```

# Create Order Table

```python
try:
    cursor.execute('''
        Create table orders (
            order_id integer primary key autoincrement,
            order_status text,
            deliver_location text,
            order_date, -- yyyy-mm-dd@  item_quantity integer,
            price real,
            customer_id integer,
            staff_id integer,
            foreign key (customer_id) references customers(customer_id),
            foreign key (staff_id) references Staff(staff_id)
        )
    ''')
    print('Order table created.')
except sqlite3.OperationalError as e:
    print(e)
```
