```python
import csv
```

```python
data = [
    
    ['Name', 'Destination', 'Price'],
    ['Nepal', 'Mustang', '$3000'],
    ['Thailand', 'Phuke', '$5000'],
    ['Italy', 'Rome', '$7000']
    ]
```

```python
#mOpen(filename, Operation(R, w, a), newline)
with open('countries.csv', 'w', newline='') as file:
    abc = csv.writer(file)
    abc.writerows(data)
    
```

# Read CSV file

```python
with open('countries.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
    
```

# CSV-Dictionary Data

# Create CSV File

```python
product_data = [
    {
        'product_name':'Bottle',
        'price':'Rs. 300',
        'color':'red'
    },
    {
        'product_name':'bowl',
        'price':'Rs. 300',
        'color':'velvet'
    },
    {
        'product_name':'ceramics',
        'price':'Rs. 300',
        'color':'blue'
    }
]
            
        

     
```

```python
with open('products.csv', 'w', newline='') as file:
    columns = [ 'product_name','price',  'color']
    writer= csv.DictWriter(file, fieldnames = columns) # must be capital D and W for DictWriter function
    writer.writeheader()
    writer.writerows(product_data)
```

```python
with open('products.csv', 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        print(row)
        for key, val in row.items():
            print(f'{key}: {val}')
        print('\n')
```

```python
new_data = [
    {
        'product_name':'jug',
        'price':'Rs. 300',
        'color':'green'
    }
]
```

```python
with open('products.csv', 'a', newline='') as file:
    columns = [ 'product_name','price',  'color']
    writer= csv.DictWriter(file, fieldnames = columns) # must be capital D and W for DictWriter function
    
    writer.writerows(new_data)
   
```

# JSON (Javascript Object Notation)

```python
# To create/write File (dump) and to read the file (Load)
```

```python
import json
```

```python
product_data = [
    {
        'product_name':'Bottle',
        'price':'Rs. 300',
        'color':'red'
    },
    {
        'product_name':'bowl',
        'price':'Rs. 300',
        'color':'velvet'
    },
    {
        'product_name':'ceramics',
        'price':'Rs. 300',
        'color':'blue'
    }
]
```

```python
# dump(data, filename, indent(spacing)
with open('product.json', 'w') as file:
    json.dump(product_data, file, indent=4)
```

```python
with open('product.json','r')as file:
    data = json.load(file)
    for row in data:
        for key,val in row.items():
            print(f'{key}:{val}')
        print('\n')
```

```python
datanew =    {
        'product_name':'VASE',
        'price':'Rs. 300',
        'color':'rainbow'
    }
```

```python
with open('product.json','w')as file:
    data.append(datanew)
    json.dump(data, file, indent=4)
```
