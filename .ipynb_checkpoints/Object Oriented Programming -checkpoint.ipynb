# Inheritance

```python
'''
Rules of inheritance
--------------------------
1. If child class needs to access data of parent class then child class must inherit/extends parent class.
2. If child class inherits parent class then child class must inherit parent class's constructor. (super())
'''
```

```python
# Parent class
class customer:
    def __init__(self, customername, customer_id, customer_address, customer_contact): #n Constructor
        self.customername = customername
        self.customer_id = customer_id
        self.customer_address = customer_address
        self.customer_contact= customer_contact

    def customerdetails(self):
        print(f'customer Id:{self.customer_id}')
        print(f'customer Name:{self.customername}')
        print(f'customer Address:{self.customer_address}')
        print(f'customer contact:{self.customer_contact}')
    
```

```python
# child class
class orders(customer):
    def __init__(self,order_id, order_items, order_location, order_status,customername, customer_id, customer_address, customer_contact):
        # inherits parent class constructor -> super()
        super().__init__(customername, customer_id, customer_address, customer_contact)
        self.order_id= order_id
        self.order_items=order_items
        self.order_location=order_location
        self.order_status=order_status
    def orderdetails(self):
         print(f'order Id:{self.order_id}')
         print(f'order Items:{self.order_items}')
         print(f'order location:self.{self.order_location}')
         print(f'order status:{self.order_status}')
        
```

```python
o=orders(101,'Perfume','Tinkune','processing','Sonica','csb102','Nayathimi','9876543111')
```

```python
o.orderdetails()
o.customerdetails()
```

```python
# child's child class
class Items(orders):
    def __init__(self, Item_Id, Item_Name, Item_quantity, Item_Price, order_id, order_items, order_location, order_status,customername, customer_id, customer_address, customer_contact):
        # inherits parent class constructor -> super()
        super().__init__(order_id, order_items, order_location, order_status,customername, customer_id, customer_address, customer_contact)
        self.Item_Id= Item_Id
        self.Item_Name=Item_Name
        self.Item_quantity=Item_quantity
        self.Item_Price=Item_Price
    def itemdetails(self):
         print(f'Item Id:{self.Item_Id}')
         print(f'Item Name:{self.Item_Name}')
         print(f'Item_Price:self.{self.Item_quantity * self.Item_Price }')
                
```

```python
i= Items(1001, 'Watch',2, 2000,101,'Perfume','Tinkune','processing','Sonica','csb102','Nayathimi','9876543111')
```

```python
o.orderdetails()
print('\n')
o.customerdetails()
print('\n')
i.itemdetails()
```

# Encapsulation

```python
'''
Data hiding
--------------------
- Access Modifier
  1. Public (self.name=name) No underscore
  2. Private (__) (self.__name=name) Double underscore
  3. Protected (_) (self._name = name ) single underscore
```

# Private Encapsulation

```python
# Prent class
class customer:
    def __init__(self, customername, customer_id, customer_address, customer_contact): #n Constructor
        self.__customername = customername
        self.__customer_id = customer_id
        self.__customer_address = customer_address
        self.__customer_contact= customer_contact

    def customerdetails(self):
        print(f'customer Id:{self.__customer_id}')
        print(f'customer Name:{self.__customername}')
        print(f'customer Address:{self.__customer_address}')
        print(f'customer contact:{self.__customer_contact}')
```

```python
# child class
class orders(customer):
    def __init__(self,order_id, order_items, order_location, order_status,customername, customer_id, customer_address, customer_contact):
        # inherits parent class constructor -> super()
        super().__init__(customername, customer_id, customer_address, customer_contact)
        self.__order_id= order_id
        self.__order_items=order_items
        self.__order_location=order_location
        self.__order_status=order_status
    def orderdetails(self):
         print(f'order Id:{self.__order_id}') # Private banako
         print(f'order Items:{self.__order_items}')
         print(f'order location:self.__{self.__order_location}')
         print(f'order status:{self.__order_status}')
        
```

```python
o=orders(101,'Perfume','Tinkune','processing','Sonica','csb102','Nayathimi','9876543111')
```

```python
o.orderdetails()
o.customerdetails()
```

# Protected Encapsulation

```python
'''
Protected
------------------
If we use protected then use those function or variables as it is private access modifiers.
'''
```

```python
# Prent class
class customer:
    def __init__(self, customername, customer_id, customer_address, customer_contact): #n Constructor
        self._customername = customername
        self._customer_id = customer_id
        self._customer_address = customer_address
        self._customer_contact= customer_contact

    def customerdetails(self):
        print(f'customer Id:{self._customer_id}')
        print(f'customer Name:{self._customername}')
        print(f'customer Address:{self._customer_address}')
        print(f'customer contact:{self._customer_contact}')
```

```python
# child class
class orders(customer):
    def __init__(self,order_id, order_items, order_location, order_status,customername, customer_id, customer_address, customer_contact):
        # inherits parent class constructor -> super()
        super().__init__(customername, customer_id, customer_address, customer_contact)
        self._order_id= order_id
        self._order_items=order_items
        self._order_location=order_location
        self._order_status=order_status
    def orderdetails(self):
         print(f'order Id:{self._order_id}') # Protected banako, user dont know kun laguage ma code lekheko chha and tyo thaa vaehalyo vaneni, private,public k ho thaa hunna
         print(f'order Items:{self._order_items}')
         print(f'order location:self._{self._order_location}')
         print(f'order status:{self._order_status}')
```

```python
o=orders(101,'Perfume','Tinkune','processing','Sonica','csb102','Nayathimi','9876543111')
```

```python
o.orderdetails()
o.customerdetails()
```
