# Simple print the alphabets as it is

```python
alphabets =['a', 'b', 'c', 'd', 'e', 'f', 'g']
alphabets
```

# unpacking list

```python
print(*alphabets)
```

# For-each loop

```python
for data in alphabets:
    print(data)
```

# Enumerate
# -> Loop both index and values of list.
# Returns 2 values -> index, value

```python
for idx, val in enumerate(alphabets): # Basically shows the index jun chai hamile herna mildaina normal code ma, enumerate use nagari we cant print the index with data
     print(f'{idx}: {val}')
```

```python
multi =[
    [1, 2, 3],
    [2,3,4,5],
    [4,4,4,4]
]
```

```python
for row in multi:
    print(row)
    for column in row:
        print(column)
```

```python
for row in multi:
    for column in row:
        print(column)
```

```python
for r_idx, row in enumerate(multi):
    print(f'\n{r_idx}:{row}')
    for c_idx, column in enumerate(row):
        print(f'\n{c_idx}:{column}')
```

# Slicing

```python
# Slice [start, stop, step], gives the value untill 4 only, 0 1 2 3 4
alphabets[0:5:1]
```

```python
# To fetch the last letter, reverse order ma -1 and to fetch first letter 0
alphabets[-1]
```

```python
alphabets[0]
```

```python
alphabets[::-1]
```

```python
alphabets[:-4:-1] # start garda empty chha, stop ma value chha -4 so d ma stop huncha, and then -1 steps le count hncha agadi, empty cha vane by default value agadiko rakhcha at start as well as end ma pani
```

```python
Letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
Letters
```

```python
Letters[-6:-3:]
```

# Add, Update, Delete

```python
Fruits=['apple','banana'] #quotation multiple halda chuttai index ma bascha
```

# Add data

```python
'''
1. append (value)
2. insert (index, value)
'''
```

```python
Fruits.append('orange') # add garcha list ma, jati choti run garyo tete time store huncha
Fruits
```

```python
for idx, data in enumerate(Fruits):
    print(f'{idx}: {data}')
```

```python
Fruits.insert(3, 'watermelon') #push garcha 3rd place of index and tya data halcha
Fruits
```

```python
print(*Fruits)
```

```python
for data in Fruits:
    print(data)
```

```python
Fruits[3] = 'Mango'
Fruits
```

# Delete Data

```python
'''
1. Remove (value) value anushar remive huncha
2. Del list(idx) index anushar remove huncha
3. Pop() --Last bata value remove garcha, LIFO, Last in First Out
'''
```

```python
Fruits.remove('watermelon')
```

```python
del Fruits[3]
```

```python
Fruits.pop()
```
