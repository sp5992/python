```python
try:
    file_name = input('enter your file name')
    with open(f'{file_name}.txt','r') as file:
        content = file.read()
        print(content)
except FileNotFoundError:
    print(f'Error: File name "{file_name}" not found in given path!')
  
```

```python
'''
Finally le chai program end vaesakesi euta msg dekhaune kaam garcha
either program run hoss or na hoss msg chai show garcha
```

```python
try:
    file_name = input('enter your file name')
    with open(f'{file_name}.txt','r') as file:
        content = file.read()
        print(content)
except FileNotFoundError:
    print(f'Error: File name "{file_name}" not found in given path!')
else:
    print(file_name)
finally:
    print('\nProgram ended successfully')
```

# Custom exception

```python
age = 15
if age >= 18:
    print('you are eligible to vote')
else:
    print('You are not')
```

```python
class AgeError(Exception):
    pass
```

```python
def checkAge(age):
    if age >= 18:
        print('You are eligible to vote')
    else:
        raise AgeError(f'Age {age} is not eligible to vote')
```

```python
checkAge(19)
```

```python
checkAge(11)
```

```python
try:
    age = int(input('enter age'))
    checkAge(age)
except AgeError as e:
    print(f'error:{e}')
except ValueError:
    print('error: only integer is accepted')
    
    
```
